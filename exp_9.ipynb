Tools for EDA - Study of Numpy Library

```python
#Declaration of Numpy

import numpy as np

a = np.array([1,10,20,30,40,50,500])
b = np.array([[1,2,3],[4,5,6]])

print(a)
print(b)

#Dimension of Array

print(a.ndim)
print(b.ndim)

#Shape of Array

print(a.shape)
print(b.shape)

#Datatype

print(a.dtype)
print(b.dtype)

#In-Built Functions using Numpy

print(np.zeros((2,3)))

print(np.ones((3,3)))

print(np.eye((3)))

print((np.arange(1,10,2)))

print((np.linspace(0,1,4)))

#Operations using Numpy

print(a+5)
print(b*2)

print(np.mean(a))

print(np.median(a))

print(np.min(a))

print(np.max(a))

print(np.sum(a))
```
